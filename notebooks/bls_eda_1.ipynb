```python
import pandas as pd
pd.set_option("display.max_columns", 50)

PATH = "/Users/ekselan/Desktop/LAMBDA/LABS/data/BLS 2019 Data - All May 2019 Data.csv"

df = pd.read_csv(PATH)

states = []

for x in df["area_title"]:
    state = x[-2:]
    states.append(state)

df["state"] = states

print(df.shape)
df.head()
```

```python
bls = df.copy()

bls = bls.dropna(axis=1)

print(bls.shape)
bls.head()
```

## Cleanup `loc_quotient`

```python
float(bls["loc_quotient"][0])
```

```python
# Change dtypes to float
bls["loc_quotient"] = bls["loc_quotient"].replace("**","0")
bls["jobs_1000"] = bls["jobs_1000"].replace("**","0")
bls["loc_quotient"] = [float(x) for x in bls["loc_quotient"]]
bls["jobs_1000"] = [float(x) for x in bls["jobs_1000"]]
```

```python
# Create subset

bls_sub = bls[["area_title","occ_title","jobs_1000","loc_quotient"]]
print(bls_sub.shape)
bls_sub.head()
```

```python
bls_sub = bls_sub.sort_values(["area_title","loc_quotient"], ascending=False)
print(bls_sub.shape)
bls_sub.head()
```

```python
cities = []
states = []

for row in bls_sub["area_title"]:
    row = row.split(",")
    cities.append(row[0])
    states.append(row[1])
    
bls_sub["city"] = cities
bls_sub["state"] = states

print(bls_sub.shape)
bls_sub.head()
```
