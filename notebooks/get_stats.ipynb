```python
# First Pass

# get_stats function(city, stat, metric=True)
def get_stats(city: str, state: str, stat: str, metric=True) -> pd.DataFrame:
    # loads in desired DataFrame
    # looks at query and matches query to appropriate csv file
    DataFrame = pd.read_csv(f"../data/weather/{city}_{state}.csv")
    
    # fixes datetime format
    DataFrame.date_time = DataFrame.date_time.apply(lambda d: datetime.strptime(d, '%Y-%m-%d %H:%M:%S'))
    
    # pulls Series for desired metric out of DataFrame
    series = DataFrame[stat]
    series.index = DataFrame.date_time
    
    # creates subsetted Series for each month in the data
    
    subsets = []
    
    # checks length of DataFrame index to see if it is a long or short dataset
    # uses splits array
    
    if len(series.index) == 16440:
        for j in range(len(splits[3:-3]) - 1):
            subset = series[series.index > splits[j]]
            subset = subset[subset.index < splits[j + 1]]
            subsets.append(subset)
            
        subsets = subsets[2:]
        
    elif len(series.index) == 17044 or len(series.index) == 17056:
        for j in range(len(splits[1:-2]) - 1):
            subset = series[series.index > splits[j]]
            subset = series[subset.index < splits[j + 1]]
            subsets.append(subset)
            
    # calculates stats for series
    
    tables = []
    time = []
    
    for sub in subsets:
        tables.append(sub.describe()[1:])
        time.append(datetime(sub.index[0].year, sub.index[0].month, sub.index[0].day))
    
    # populates DataFrame with those stats
    
    minimum = []
    mean = []
    med = []
    maximum = []

    for i in range(len(tables)):
        mean.append(tables[i]["mean"])
        minimum.append(tables[i]["min"])
        med.append(tables[i]["50%"])
        maximum.append(tables[i]["max"])

    data = {
        "month" : time,
        "mean" : mean,
        "min" : minimum,
        "med" : med,
        "max" : maximum,
    }
    
    df = pd.DataFrame(data)
    
    # resets new DataFrame index to the datetime of the first entry
    df.set_index("month", inplace=True)
    
    # sets freq of df's index to "MS" for monthly
    df.index.freq = "MS"
    
    return df

# Second Pass
# if metric=False
# convert values at stats stage to imperial units

# Third Pass

# populate database with historical data <- can be stored in another function which is called if the data are not already there,
#                                           or are out-of-date

# build if then loop to query database for information
# if information present:
    # return information

# if information absent:
    # calculate info.
```

```python
def get_min(series, rule):
    return series.resample(rule=rule).min()
    
def get_mean(series, rule):
    return series.resample(rule=rule).mean()
    
def get_med(series, rule):
    return series.resample(rule=rule).med()
    
def get_max(series, rule):
    return series.resample(rule=rule).max()

def get_stats(city: str, state: str, metric: str, stat: str) -> pd.Series:
    # loads in desired DataFrame
    # looks at query and matches query to appropriate csv file
    DataFrame = pd.read_csv(f"../data/weather/{city}_{state}.csv")
    
    # fixes datetime format
    DataFrame.date_time = DataFrame.date_time.apply(lambda d: datetime.strptime(d, '%Y-%m-%d %H:%M:%S'))
    
    # pulls Series for desired metric out of DataFrame
    series = DataFrame[metric]
    series.index = DataFrame.date_time
    
    if stat == "min":
        return get_min(series, rule="MS").to_json()
        
    elif stat == "mean":
        return get_mean(series, rule="MS").to_json()
        
    elif stat == "med":
        return get_med(series, rule="MS").to_json()
        
    elif stat == "max":
        return get_max(series, rule="MS").to_json()
```
