```python
# transposition function

def transpose(city: str) -> pd.DataFrame:
    DataFrame = df[df["City"] == city].T
    DataFrame.columns = DataFrame.loc["bedroom_size"].values
    city_name = DataFrame["Studio"]["City"]
    DataFrame.drop(["City", "bedroom_size"], inplace=True)
    DataFrame.index = pd.to_datetime(DataFrame.index, format="price_%Y_%m")
    DataFrame.index.freq = "MS"
    DataFrame.reset_index(inplace=True)
    DataFrame.insert(0, "city", [city_name[:-4]] * len(DataFrame.index))
    DataFrame.insert(1, "state", [city_name[-2:]] * len(DataFrame.index))
    return DataFrame
```

```python
# Recombining city and state

City = []

for i in range(len(df.index)):
    City.append(f"{df.city[i]}, {df.state[i]}")
    
df.insert(0, "City", City)
df.drop(["city", "state"], axis=1, inplace=True)

print(df.shape)
df.head()
```

```python
dfs = []

for c in list(df.City.value_counts().keys()):
    dfs.append(transpose(c))
    
len(dfs)
```

```python
dfs[0].head()
```

```python
from functools import reduce

df_merged = reduce(lambda left, right: pd.merge(left, right, on=list(dfs[0].columns), how='outer'), dfs)

print(df_merged.shape)
df_merged.head()
```
