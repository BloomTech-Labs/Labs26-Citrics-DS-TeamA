```python
# Dropping NaNs for now in the interest of fast-tracking
# an endpoint for FE team
df = df.dropna()

print(df.shape)
df.head()
```

```python
df.dtypes

# Strange ... some of the prices are ints, some are floats
# Will need to correct that in case it causes headaches for FE team
```

```python
# Can check if dtype is float, and if so, change its
# values to integers

for col in df.columns.to_list():
    if df[col].dtype == float:
        df[col] = [int(x) for x in df[col].to_list()]           
```

```python
# Also want to remove the state from the city names

cities = []

for x in df["location"]:
    city = x[:-2]
    city = city.replace(",","")
    city = city.strip()
    
    cities.append(city)

df["city"] = cities

print(df.shape)
df.head()
```
